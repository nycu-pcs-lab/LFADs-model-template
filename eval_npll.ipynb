```python
res_file_path = "./tb_data/csim_results.log"
input_file_path = "./tb_data/tb_input_features.dat"
keras_out_file_path = "./tb_data/tb_output_predictions.dat"
cosim_file_path = "./tb_data/rtl_cosim_results.log"
```

```python
import numpy as np
```

```python
x_test = np.loadtxt(input_file_path)
keras_pred = np.loadtxt(keras_out_file_path)
```

```python
print("x_test shape: ",x_test.shape)
print("keras_pred shape: ",keras_pred.shape)
```

```python
x_test = x_test[0:5110]
```

```python
x_test = x_test.reshape(1,73,70)
keras_pred = keras_pred.reshape(1,73,70)
```

```python
csim_res_f = open(res_file_path)
csim_res = np.array(csim_res_f.read().split()).reshape(1,73,70)
```

```python
#cosim_res_f = open(cosim_file_path)
#cosim_res = np.array(cosim_res_f.read().split()).reshape(1,73,70)
```

```python
print("csim result shape: ",csim_res.shape)
```

```python
import tensorflow as tf
```

```python
def evaluate_NPLL(targets, pred_logrates):
    targets = tf.cast(targets, dtype=tf.float32)
    logrates = tf.cast(tf.math.log(0.01) + pred_logrates, tf.float32)  # Timestep
    npll = tf.nn.log_poisson_loss(targets=targets,log_input=logrates, compute_full_loss=True)
    results = tf.reduce_sum(npll, axis=[1, 2]) # sum up each batch seperately
    results = tf.reduce_mean(results) # batch mean

    return results.numpy() # negative possion loglikelihood
```

```python
print("keras npll result: ",evaluate_NPLL(targets=x_test,pred_logrates=keras_pred))
print("csim npll result: ",evaluate_NPLL(targets=x_test,pred_logrates=csim_res))
#print("cosim result: ",evaluate_NPLL(targets=x_test,pred_logrates=cosim_res))
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
print("csim mse error: ",mean_absolute_error(keras_pred.flatten(),csim_res.flatten()))
```
